```python
#importamos la libreria de pandas
import pandas as pd
# leemos nuestro archivo csv y lo guardamos en un data frame
df = pd.read_csv(r"C:\Users\marcos.sandovalr\Documents\Marcos Privada\Python Big Data\CarrosRobados.csv", nrows=1000)
```

```python
#df.dtypes
#convertimos la columna DateStolen la cual era tipo object en tipo fecha datetime, para esto utilizamos el metodo de pandas
#pd.to_datetime al cual le pasamos como parametro la columna a convertir y definimos el formato de la fecha
df['DateStolen'] = pd.to_datetime(df['DateStolen'], format="%d/%m/%Y")
#crear una columna nueva a partir de las condiciones de otras
df["AñoRobo"] =  df["DateStolen"].dt.year
df["MesRobo"] = df["DateStolen"].dt.month
```
