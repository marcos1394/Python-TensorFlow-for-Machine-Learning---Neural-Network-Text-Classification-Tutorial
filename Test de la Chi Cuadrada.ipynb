```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\marcos.sandovalr\Documents\Marcos Privada\Python Big Data\Python-TensorFlow-for-Machine-Learning-Neural-Network-Text-Classification-Tutorial\Eventos-US.xlsx")
df['Duracion Evento'] =  (df['Fecha Fin'] - df['Fecha Inicio'])
df.head(20)
```

```python
np.random.seed(0)
df = df[df["Desastre"].isin(["Drought","Tropical Cyclone","Flooding"])]
df = df.sample(frac=1)
```

```python
df["Catastrofe"] = df["Muertes"] > 20
observados = pd.crosstab(index=df['Catastrofe'], columns=df['Desastre'], margins=True)
observados
```

```python
from scipy.stats import chi2_contingency
```

```python
test = chi2_contingency(observados)
test
```

```python
esperados = pd.DataFrame(test[3])
esperados
```

```python
esperados_rel = round(esperados.apply(lambda r: r/len(df)*100,axis=1),2)
```

```python
observados_rel = round(observados.apply(lambda r: r/len(df)*100,axis=1),2)
```
