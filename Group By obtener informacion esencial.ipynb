```python
import pandas as pd
df = pd.read_csv(r"C:\Users\marcos.sandovalr\Documents\Marcos Privada\Python Big Data\CarrosRobados.csv", nrows=1000)
df.head()
df['DateStolen'] = pd.to_datetime(df['DateStolen'], format="%d/%m/%Y")
df["AñoRobo"] = df["DateStolen"].dt.year
df["MesRobo"] = df["DateStolen"].dt.month
df["DiaRobo"] = df["DateStolen"].dt.day
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo max estamos indicando nos agrupe los modelos
#de carros por el año maximo con el que aparecen
df.groupby(by="VehicleModel")["ModelYear"].max()
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo mean estamos indicando nos agrupe los modelos
#de carros por el año promedio
df.groupby(by="VehicleModel")["ModelYear"].mean()
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo min estamos indicando nos agrupe los modelos
#de carros por el año menor
df.groupby(by="VehicleModel")["ModelYear"].min()
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo count estamos indicando nos agrupe los modelos
#de carros y nos cuente el número que hubo de años que tienen
df.groupby(by="VehicleModel")["ModelYear"].count()
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo describe estamos indicando nos agrupe los modelos
#de carros y nos de un resumen estadistico por año
df.groupby(by="VehicleModel")["ModelYear"].describe()
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo mean estamos indicando nos agrupe los modelos
#de carros y nos da el promedio en este caso sera para dos columnas el modelo y el año de robo 
df.groupby(by="VehicleModel")[["ModelYear","AñoRobo"]].mean() 
```

```python
#el metodo group by nos permite agrupar datos, asignando al parametro by la columna por la que queremos agrupar
#y la segunda columna es la que condiciona la agrupacion en este caso con el metodo mean estamos indicando nos agrupe los modelos
#de carros por el año del carro y el año de robo, en esta operacion utilizaremos dos objetos group by realizando una operacion
#aritmetica en este caso una resta
df.groupby(by="VehicleModel")[["ModelYear","AñoRobo"]].max() - df.groupby(by="VehicleModel")[["ModelYear","AñoRobo"]].min()
```
